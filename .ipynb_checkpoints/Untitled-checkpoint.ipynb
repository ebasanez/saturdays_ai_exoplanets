```python
import pandas as pd
import lightkurve as lk
from lightkurve import LightCurveCollection
from lightkurve import search_lightcurvefile    
```

```python
(koi_kic,t0,period,mission) = (10666592,121.358542,2.204735,'Kepler')
#(koi_kic,t0,period) = (10797460,170.538750,9.488036)
```

```python
lcs = search_lightcurvefile('KIC ' + str(koi_kic)).download_all(quality_bitmask='hardest').PDCSAP_FLUX
```

```python
lcs = LightCurveCollection([lc for lc in lcs if lc.mission == mission])
lc = lcs.stitch().flatten()
```

```python
lc.fold(period, t0).plot()
```

```python
df = pd.DataFrame({'time' : lc.time, 'flux' : lc.flux})
df.head()
```

```python
df['period'] = df['time'].map(lambda t: int((t - t0 - period/2) // period) )
print(df.head())
df.tail()
```

```python
# Remove first and last period (could be incomplete)
max_period = df.period.max()
df = df[df.period != -1]
df = df[df.period != max_period]
df = df.reset_index(drop=True)

```

```python
dfs = []
for i in range(max_period - 1):
    dfs.append(df[['time','flux']][df.period == i].reset_index(drop=True))
dfs[1].head()
```
