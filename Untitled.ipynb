```python
import numpy as np
import pandas as pd
import lightkurve as lk
from lightkurve import LightCurveCollection
from lightkurve import search_lightcurvefile    
from es.saturdaysai.exoplanets.lightKurveApi.lightKurveClient import LightKurveClient 
```

```python
(koi_kic,t0,period,duration,mission) = (10666592,121.358542,2.204735,0.162027,'Kepler')
#(koi_kic,t0,period) = (10797460,170.538750,9.488036)
```

```python
lcs = search_lightcurvefile('KIC ' + str(koi_kic)).download_all(quality_bitmask='hardest').PDCSAP_FLUX
```

```python
lcs = LightCurveCollection([lc for lc in lcs if lc.mission == mission])
lc_flat = lcs.stitch().flatten()
```

```python
lc_fold = lc_flat.fold(period, t0)
lc_fold.plot()
```

```python
local_view_witdh =  4
phased_duration = duration / period
transit_mask = (lc_fold.phase > - local_view_witdh * phased_duration) & (lc_fold.phase < local_view_witdh * phased_duration)
lc_local = lc_fold[transit_mask]
lc_local.scatter()
```

```python
binsize =len(lc_local.flux)/201
binsize
```

```python
lc_local_bin = lc_local.bin(binsize = binsize, method = 'median').normalize() - 1
lc_local_bin = (lc_local_bin / np.abs(lc_local_bin.flux.min()) ) * 2.0 + 1
lc_local_bin.scatter()
lc_local_bin.flux.shape
```

```python
t_max = lc_flat.time.max()
t_period_init = t0 - period / 2
t_period_end = t_period_init + period
shown = 0
lc_period_folds = []
while t_period_end < t_max:
    t_period_init += period
    t_period_end += period
    period_mask = ((lc_flat.time > t_period_init) & (lc_flat.time < t_period_end))
    lc_period_fold = lc_flat[period_mask].fold(period = period, t0 = (t_period_init + period / 2))
    lc_period_folds.append(lc_period_fold)    
len(lc_period_folds)
```
